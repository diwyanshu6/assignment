```python
# m=sum[((x-X)*(y-Y))/(x-X)**2] these value came from partial diffrention of squared error wrt m and b  equated to zero to minimise error
# X=mean of X
```

```python
class ScratchLR:
  def __init__(self,m=0,b=0):
    self.m=m
    self.b=b
  def fit(self,x,y):
    x_mean=np.mean(x)
    y_mean=np.mean(y)
    numerator=0
    denominator=0
    for i in range(len(x)):
      numerator=numerator+(x[i]-x_mean)*(y[i]-y_mean)
      denominator=denominator+(x[i]-x_mean)**2
    self.m=numerator/denominator
    self.b=y_mean-(self.m*x_mean)
    return self.m,self.b
  def prediction(self,x_test):
    y_pred=self.m*x_test+self.b
    return y_pred
```
